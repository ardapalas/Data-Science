```python
df.dropna(subset=["Current Ver","Android Ver", "Content Rating", "Type"],inplace=True)
```

```python
rating_mode = df["Rating"].mode()[0]
print(rating_mode)
```

```python
df.isna().sum()
```

```python
df["Rating"] = df["Rating"].fillna(rating_mode)
```

```python
df.head()
```

```python
df.info()
```

```python
#df[~df["Reviews"].str.isnumeric()]
#df = df.drop(df.index[123])
```

```python
df["Reviews"] = df["Reviews"].astype(int)
```

```python
df.describe()
```

```python
df["Size"].unique()
```

```python
df["Size"] = df["Size"].str.replace("k","")
df["Size"] = df["Size"].str.replace("M","000")
df["Size"] = df["Size"].replace("Varies with device",np.nan)
```

```python
df["Size"] = df["Size"].astype(float)
```

```python
df["Size"]
```

```python
df.info()
```

```python
df["Installs"].value_counts()
```

```python
df["Installs"] = df["Installs"].str.replace("+","")
df["Installs"] = df["Installs"].str.replace(",","")
```

```python
df["Installs"].value_counts()
```

```python
df["Installs"] = df["Installs"].astype(int)
```

```python
df.info()
```

```python
df["Price"].unique()
```

```python
df["Price"] = df["Price"].str.replace("$","")
df["Price"] = df["Price"].astype(float)
```

```python
df["Price"].value_counts()
```

```python
df.info()
```

```python
df.head()
```

```python
df["Last Updated"]
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
```

```python
df["Last Updated"]
```

```python
df["Day"] = df["Last Updated"].dt.day
df["Month"] = df["Last Updated"].dt.month
df["Year"] = df["Last Updated"].dt.year
```

```python
df[df.duplicated("App")].shape
```

```python
df = df.drop_duplicates(subset='App', keep='first')
```

```python
df.info()
```

```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
print(numerical_features,categorical_features)
```

```python
plt.figure(figsize=(15,10))

for i in range(len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df[numerical_features[i]], color = "red", fill=True)
    plt.xlabel(numerical_features[i])

plt.show()
```
